```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt
w=np.logspace(1,3,10000)
z=np.exp(w/100)*1j+10000
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
ax.plot(np.real(z),-np.imag(z),np.log(w))
# big=np.max(np.real(z),np.imag(z))
# ax.set_aspect("equal")
ax.set_ylim(0,10)
```
